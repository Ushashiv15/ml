```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def es_co(x,y):
    n=np.size(x)
    mx=np.mean(x)
    my=np.mean(y)
    sxy=np.sum((x-mx)*(y-my))
    sx=np.sum((x-mx)**2)
    b1=sxy/sx
    b0=my-b1*mx
    return(b0,b1)
```

```python
def plots(x,y,b):
    plt.scatter(x,y)
    yp=b[0]+b[1]*x
    plt.plot(x,yp,color='red')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()
x=np.array([0,1,2,3,5,6,7,8,9])
y=np.array([1,3,2,5,7,8,8,9,10])
a=es_co(x,y)
print('estimated coefficient b0 & b1 are: ',a)
plots(x,y,a)
```
